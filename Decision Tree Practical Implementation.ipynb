```python
 df = sns.load_dataset('iris')
```

### Independent Feature and Dependent Feature

```python
X = df.iloc[:,:-1]
Y = iris.target
```

```python
X,Y
```

### Train Test and Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.33,random_state = 42)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

## Preprunning

```python
model = DecisionTreeClassifier(max_depth = 3)
parameter = {"criterion" : ["gini", "entropy"],"max_depth": [2,3,4,5,6],
    "min_samples_split": [2,5,10],
    "min_samples_leaf": [1,2,4]
}
```

```python
grid = GridSearchCV(estimator = DecisionTreeClassifier(), param_grid=parameter,cv=5,scoring="accuracy")
grid
```

```python
grid.fit(X_train,Y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
best_model = grid.best_estimator_
```

```python
Y_pred = best_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(Y_test,Y_pred)
```

```python
confusion_matrix(Y_test,Y_pred) #actual(horizontal) vs predicted(vertical) (targetnames)
```

```python
classification_report(Y_test,Y_pred)
```

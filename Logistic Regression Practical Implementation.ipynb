```python
import pandas as pd 
import numpy as np 
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['species'].unique()
```

```python
df['species'] != 'setosa'
```

```python
df = df[df['species'] != 'setosa']
```

```python
df.head()
```

```python
df['species'] = df['species'].map({'versicolor' : 0,'virginica' : 1})
```

```python
df.head()
```

### Splitting dataset into independent and dependent features 

```python
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
parameter = {'penalty':['l1','l2','elasticnet',None],'C': [1,2,3,4,10,20,30,40,50,60],'max_iter':[100,200,300]}
```

```python
classifier_reg = GridSearchCV(classifier,param_grid = parameter,scoring = 'accuracy',cv = 5)
```

```python
classifier_reg.fit(X_train,Y_train)
```

```python
print(classifier_reg.best_params_)
```

```python
print(classifier_reg.best_score_)
```

```python
y_pred = classifier_reg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, y_pred)
```

### Classification Report

```python
from sklearn.metrics import classification_report
classification_report(y_pred,Y_test)
```
